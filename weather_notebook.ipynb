```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import datetime

import openweathermapy as owm
# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=3000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=3000)
lat_lngs = zip(lats, lngs)

# Lets plot the random locations to verify they are evenly spread out
plt.scatter(lats, lngs, s=5)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show
```

```python
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print('We have coordinates for ' +str(len(cities))+ ' cities.')
```

```python
# Lets peek into the city list to verify it looks like a list of city names
cities[:10]
```

```python
# Define empty lists to append weather data
weather_data = []
name = []
date = []
latitude = []
longitude = []
clouds = []
wind = []
humidity = []
pressure = []
temp = []
tmax = []
tmin = []
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Create a settings object with your API key and preferred units
settings = {"units": "imperial", "appid": api_key}

for city in range(len(cities)):

    try:
        weather_data.append([owm.get_current(cities[city], **settings)])
        print("Processing Record " +str(city) + ": "  + str(cities[city]))

    except:
        print("Missing Data for: " + str(cities[city]))
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Let take a look at the structure of the data coming from the API
# so we can determine how to save it into lists
weather_data[0][0]
```

```python
# Save the data into lists
for city in range(len(weather_data)):
    name.append(weather_data[city][0]['name'])
    date.append(weather_data[city][0]['dt']) 
    latitude.append(weather_data[city][0]['coord']['lat'])
    longitude.append(weather_data[city][0]['coord']['lon'])
    clouds.append(weather_data[city][0]['clouds']['all'])
    wind.append(weather_data[city][0]['wind']['speed'])
    humidity.append(weather_data[city][0]['main']['humidity'])
    pressure.append(weather_data[city][0]['main']['pressure'])
    tmax.append(weather_data[city][0]['main']['temp_max'])
```

```python
# Create a dataframe with all data
weather_df = pd.DataFrame({'City':name,'Date':date,
                           'Latitude':latitude,'Longitude':longitude,
                           'Cloudiness':clouds,'Wind Speed':wind,
                           'Humidity':humidity,'Pressure':pressure,'T_Max':tmax})
```

```python
# Lets take a look at the dataframe, to check data looks fine
weather_df.head()
```

```python
# Save dataframe into output csv file
weather_df.to_csv(output_data_file, index=False) # defined in the first cell: cities.csv
```
